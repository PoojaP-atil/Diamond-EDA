```python
df
```

The diamonds data set is a simple dataset that consists of observation:
price price in US dollars (\$326 — \$18,823)

carat weight of the diamond (0.2–5.01)

cut quality of the cut (Fair, Good, Very Good, Premium, Ideal)

color diamond colour, from J (worst) to D (best)

clarity a measurement of how clear the diamond is (I1 (worst), SI2, SI1, VS2, VS1, VVS2, VVS1, IF (best))

x length in mm (0–10.74)

y width in mm (0–58.9)

z depth in mm (0–31.8)

depth total depth percentage = z / mean(x, y) = 2 * z / (x + y) (43–79)

table width of top of diamond relative to widest point (43–95)

## Step 2: Feature Assessment and Visualization

After having a peek at the overall data descriptors, we need to zoom in on our dataset’s features, in order to get some insights on their individual properties — Univariate Analysis — as well their interactions and relationships — Multivariate Analysis.

### Univariate Analysis
Analyzing the individual characteristics of each feature is crucial as it will help us decide on their relevance for the analysis and the type of data preparation they may require to achieve optimal results.

1. Scatterplot : A scatter plot is a great way to visualize the relationship between two continuous variables.

```python
sns.scatterplot(x='carat', y='price', hue='cut', size='depth', data=df)
plt.title('Scatter Plot of Carat vs Price (Based on cut)',color='blue',fontsize=16, fontweight='bold', fontname='Arial')
```

```python
sns.scatterplot(x='carat', y='price', hue='color', size='depth', data=df)
plt.title('Scatter Plot of Carat vs Price (Based on color)',color='blue',fontsize=16, fontweight='bold', fontname='Arial')
```

```python
sns.scatterplot(x='carat', y='price', hue='clarity', size='depth', data=df)
plt.title('Scatter Plot of Carat vs Price (Based on clarity)',color='blue',fontsize=16, fontweight='bold', fontname='Arial')
```

2. Stripplot : A strip plot is a type of categorical scatter plot that represents data points as individual dots along one axis, typically the x-axis. Each dot corresponds to an observation in the dataset.

```python
sns.stripplot(x='cut', y='carat', data=df,jitter=True,marker='D',palette='pastel')
plt.title('Strip Plot of Carat by Cut',color='orange',fontsize=16, fontweight='bold', fontname='Arial')
```

```python
sns.stripplot(x='color', y='carat', data=df,jitter=True,marker='o',palette='deep')
plt.title('Strip Plot of Carat by Color',color='orange',fontsize=16, fontweight='bold', fontname='Arial')
```

```python
sns.stripplot(x='clarity', y='carat', data=df,jitter=True,marker='o',palette='bright')
plt.title('Strip Plot of Carat by Clarity',color='orange',fontsize=16, fontweight='bold', fontname='Arial')
```

```python
sns.stripplot(x='cut', y='price', data=df,jitter=True,palette='colorblind')
plt.title('Strip Plot of Price by Cut',color='orange',fontsize=16, fontweight='bold', fontname='Arial')
```

```python
sns.stripplot(x='color', y='price', data=df,jitter=True,palette='flare')
plt.title('Strip Plot of Price by Color',color='orange',fontsize=16, fontweight='bold', fontname='Arial')
```

```python
sns.stripplot(x='clarity', y='price', data=df,jitter=True,palette='rocket')
plt.title('Strip Plot of Price by Clarity',color='orange',fontsize=16, fontweight='bold', fontname='Arial')
```

3. Barplot : A bar plot in Seaborn (and matplotlib) is a useful way to visualize categorical data where the height of the bars represents some aggregated value, typically a mean or count, for each category.

```python
sns.barplot(x='cut', y='price', data=df, color='Red')
plt.title('Bar Plot of Price by Cut',color='purple',fontsize=16, fontweight='bold', fontname='Arial')
```

```python
sns.barplot(x='color', y='price', data=df, color='salmon')
plt.title('Bar Plot of Price by Color',color='purple',fontsize=16, fontweight='bold', fontname='Arial')
```

```python
sns.barplot(x='clarity', y='price', data=df, color='pink')
plt.title('Bar Plot of Price by Clarity',color='purple',fontsize=16, fontweight='bold', fontname='Arial')
```

### Multivariate Analysis 
For Multivariate Analysis, best practices focus mainly on two strategies: analyzing the interactions between features, and analyzing their correlations.

4. Pairplot : A pairplot is a graph that plot pairwise relationships in a dataset.

```python
sns.pairplot(df)
```

5. Lineplot : A line plot in Seaborn (and matplotlib) is a visualization that displays data points connected by straight line segments. It is commonly used to show trends over time or to explore the relationship between two continuous variables.

```python
sns.lineplot(x='carat', y='x', data=df, color='salmon')
plt.title('Line Plot of Length by Carat',color='red',fontsize=16, fontweight='bold', fontname='Arial')
```

```python
sns.lineplot(x='carat', y='y', data=df, color='yellow')
plt.title('Line Plot of width by Carat',color='red',fontsize=16, fontweight='bold', fontname='Arial')
```

```python
sns.lineplot(x='carat', y='z', data=df, color='green')
plt.title('Line Plot of depth by Carat',color='red',fontsize=16, fontweight='bold', fontname='Arial')
```

6. Histogram : A histogram is a type of bar graph that represents the distribution of a dataset. It divides the data into bins (or intervals) and shows the frequency (or count) of data points that fall into each bin.

```python
sns.histplot(data=df, x='carat', color='black')
plt.title('The distribution of diamond carat values in the dataset.',color='green',fontsize=16, fontweight='bold', fontname='Arial')
plt.show()
```

```python
sns.histplot(data=df, x='price', color='salmon')
plt.title('The distribution of diamond price in the dataset.',color='green',fontsize=16, fontweight='bold', fontname='Arial')
plt.show()
```

```python
sns.histplot(data=df, x='depth', hue='cut', color='cyan')
plt.title('Visualizes how the depth of diamonds varies with different cut qualities.',color='green',fontsize=16, fontweight='bold', fontname='Arial')
```

```python
sns.histplot(data=df, x='table', hue='color', multiple='stack', color='red')
plt.title('Visualizes how the table of diamonds varies with different color grades.',color='green',fontsize=16, fontweight='bold', fontname='Arial')
```

```python
sns.histplot(data=df, x='x', hue='clarity', element='step', color='blue')
plt.title('Visualizes how the length of diamonds varies with clarity grades.',color='green',fontsize=16, fontweight='bold', fontname='Arial')
```

```python
sns.histplot(data=df, x='y', hue='clarity', element='poly', color='pink')
plt.title('Visualizes how the width of diamonds varies with clarity grades.',color='green',fontsize=16, fontweight='bold', fontname='Arial')
```

```python
sns.histplot(data=df, x='z', hue='cut', element='step', stat='density', color='orange')
plt.title('Visualizes how the depth of diamonds varies with cut grades.',color='green',fontsize=16, fontweight='bold', fontname='Arial')
```

7. Heatmap : A heatmap is a graphical representation of data where the values of a matrix are represented as colors. It is an effective way to visualize the magnitude of a phenomenon as color in two dimensions. Heatmaps are particularly useful for showing correlations, distributions, or densities of data points in a matrix format.

```python
new=df[['carat','depth','table','price','x','y','z']]
```

```python
sns.heatmap(data=new.corr(),annot=True)
```

8. Boxplot : A boxplot is a graphical representation of data that displays the distribution of numerical data based on quartiles. It is particularly useful for identifying outliers and comparing distributions between different categories or groups.

```python
sns.boxplot(x='carat', y='cut', data=df, color='green', width=0.5, linewidth=2.5)
plt.title('Box Plot of Cut by Carat',color='black',fontsize=16, fontweight='bold', fontname='Arial')
```

```python
sns.boxplot(x='carat', y='color', data=df, color='yellow', width=0.5, linewidth=2.5)
plt.title('Box Plot of Color by Carat',color='black',fontsize=16, fontweight='bold', fontname='Arial')
```

```python
sns.boxplot(x='carat', y='clarity', data=df, color='orange', width=0.5, linewidth=2.5)
plt.title('Box Plot of Clarity by Carat',color='black',fontsize=16, fontweight='bold', fontname='Arial')
```

9. Violinplot : A violin plot is a method of plotting numeric data and its probability density. It combines aspects of a box plot and a kernel density plot, showing the distribution of quantitative data across several levels of one (or more) categorical variables. Violin plots are useful for comparing distributions and understanding the density of data points at different values.

```python
sns.violinplot(x='cut', y='carat', data=df, color='cyan', width=1)
plt.title('Violin Plot of Carat by Cut',color='cyan',fontsize=16, fontweight='bold', fontname='Arial')
```

```python
sns.violinplot(x='color', y='carat', data=df, palette='muted', width=1)
plt.title('Violin Plot of Carat by Color',color='cyan',fontsize=16, fontweight='bold', fontname='Arial')
```

```python
sns.violinplot(x='clarity', y='carat', data=df, palette='bright', width=1)
plt.title('Violin Plot of Carat by Clarity',color='cyan',fontsize=16, fontweight='bold', fontname='Arial')
```

10. Pointplot : A point plot in Seaborn is used to show point estimates and confidence intervals as a function of one categorical variable (x) and one numeric variable (y). It's particularly useful for visualizing how the mean (or other estimator) of y changes with different levels of x.

```python
sns.pointplot(x='cut', y='carat', data=df)
plt.title('Point Plot of Carat by Cut',color='royalblue',fontsize=16, fontweight='bold', fontname='Arial')
```

```python
sns.pointplot(x='color', y='carat', data=df,color='yellow')
plt.title('Point Plot of Carat by Color',color='royalblue',fontsize=16, fontweight='bold', fontname='Arial')
```

```python
sns.pointplot(x='clarity', y='carat', data=df,color='purple')
plt.title('Point Plot of Carat by Clarity',color='royalblue',fontsize=16, fontweight='bold', fontname='Arial')
```
